# 텐서의 기본 연산

## 텐서의 노름

```python
A = [ [[1,0,1,0],[2,1,0,1],[3,4,0,2]], 
      [[3,1,2,0],[1,0,4,2],[0,1,0,2]] ]
```

```python
print(A)
```

```python
n = len(A[0])
p = len(A[0][0])
q = len(A)
```

```python
print(n)
print(p)
print(q)
```

```python
norm = 0
for i in range(0, n):
    for j in range(0, p):
        for k in range(0, q):
            norm += A[k][i][j]**2
norm = norm**0.5
```

```python
print(norm)
```

# 텐서의 내적

```python
A = [ [[1,0,1,0],[2,1,0,1],[3,4,0,2]], 
      [[3,1,2,0],[1,0,4,2],[0,1,0,2]] ]

B = [ [[2,1,3,1],[0,2,1,3],[1,0,1,1]], 
      [[1,2,3,1],[3,1,0,3],[0,2,3,1]] ]
```

```python
n = len(A[0])
p = len(A[0][0])
q = len(A)
```

```python
inner_product = 0
for i in range(0, n):
    for j in range(0, p):
        for k in range(0, q):
            inner_product += A[k][i][j]*B[k][i][j]
```

```python
print(inner_product)
```

# 텐서 곱

```python
U = [[1,3,5],[2,4,6]]

A = [ [[1,4,7,10],[2,5,8,11],[3,6,9,12]], 
      [[13,16,19,22],[14,17,20,23],[15,18,21,24]] ]
```

```python
n1 = len(U)
p1 = len(U[0])

n2 = len(A[0])
p2 = len(A[0][0])
q2 = len(A)
```

```python
print(n1)
print(p1)
```

```python
print(n2)
print(p2)
print(q2)
```

```python
tensor_product = []
for k in range(0, q2):
    sub_matrix = []
    for i in range(0, n1):
        row = []
        for j2 in range(0,p2):
            val = 0
            for j1 in range(0,p1):
                val += U[i][j1]*A[k][j1][j2]
            row.append(val)
        sub_matrix.append(row)
    tensor_product.append(sub_matrix)
```

```python
tensor_product
```
