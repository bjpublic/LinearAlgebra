# 텐서의 내적

```python
import numpy as np

A = np.array([ [[1,0,1,0],[2,1,0,1],[3,4,0,2]], 
               [[3,1,2,0],[1,0,4,2],[0,1,0,2]] ])

B = np.array([ [[2,1,3,1],[0,2,1,3],[1,0,1,1]], 
               [[1,2,3,1],[3,1,0,3],[0,2,3,1]] ])
```

```python
C = np.multiply(A, B)
print(C)
```

```python
inner_product = np.sum(C)
print(inner_product)
```

# 행렬 합치기(Concatenate)

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
```

```python
print(A)
```

```python
print(B)
```

```python
C = np.concatenate((A, B), axis=0)
print(C)
```

```python
C.shape
```

```python
D = np.concatenate((A, B), axis=1)
print(D)
```

```python
D.shape
```

```python
E = np.concatenate((A, B), axis=None)
print(E)
```

# 텐서의 행렬화

```python
A = np.array([[[1, 4, 7, 10],[2, 5, 8, 11],[3, 6, 9, 12]],
              [[13, 16, 19, 22],[14, 17, 20, 23],[15, 18, 21, 24]]])
print(A)
```

```python
A.shape
```

```python
print(A[0])
```

```python
print(A[1])
```

## mode1

```python
mode1 = np.concatenate((A[0], A[1]), axis=1)
print(mode1)
```

## mode2

```python
print(A[0])
```

```python
print(A[0].T)
```

```python
mode2 = np.concatenate((A[0].T, A[1].T), axis=1)
print(mode2)
```

## mode3

```python
A1 = A[0].T.reshape(-1)
print(A1)
```

```python
A2 = A[1].T.reshape(-1)
print(A2)
```

```python
A3 = np.concatenate((A1, A2), axis=0)
print(A3)
```

```python
mode3 = A3.reshape(2,-1)
print(mode3)
```

# 텐서 곱

```python
U = np.array([[1,3,5],[2,4,6]])

A = np.array([ [[1,4,7,10],[2,5,8,11],[3,6,9,12]], 
               [[13,16,19,22],[14,17,20,23],[15,18,21,24]] ])
```

```python
UA1 = np.matmul(U, A[0])
print(UA1)
```

```python
UA2 = np.matmul(U, A[1])
print(UA2)
```

```python
C = np.concatenate((UA1, UA2), axis=0)
print(C)
```

```python
tensor_product = C.reshape(2, 2, 4)
print(tensor_product)
```
